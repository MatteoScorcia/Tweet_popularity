```python
import pandas as pd
from matplotlib import pyplot as plt
data = pd.read_csv('../dataset/dataset.csv')
```

```python
data.head()
```

```python
data.index[0]
```

```python
plt.style.use("seaborn")
plt.title('tf-idf of tweets distribution', color='black', size=15)
plt.xlabel('Tweet index', color='black', size=15)
plt.ylabel('tf-idf', color='black', size=15)
plt.scatter(data.index.tolist(), data['tf-idf'].tolist(), alpha=0.4)
```

```python
plt.style.use("seaborn")
plt.title('ratio distribution', color='black', size=15)
plt.xlabel('Tweet index', color='black', size=15)
plt.ylabel('RATIO : (retweets / tot_followers of the page)', color='black', size=15)
plt.plot(data.index.tolist(), data['ratio'].tolist(), alpha=0.8)
```

```python
plt.style.use("seaborn")
plt.title('Urls distribution', color='black', size=15)
plt.xlabel('Tweet index', color='black', size=15)
plt.ylabel('Number of Urls', color='black', size=15)
plt.scatter(data.index.tolist(), data['urls'].tolist(), alpha=0.8, s=15)
```

```python
plt.style.use("seaborn")
plt.title('Retweets distribution', color='black', size=15)
plt.xlabel('Tweet index', color='black', size=15)
plt.ylabel('Number of retweets', color='black', size=15)
plt.plot(data.index.tolist(), data['retweet_count'].tolist(), alpha=0.8)
```

```python
plt.style.use("seaborn")
plt.title('Publishing time distribution', color='black', size=15)
plt.xlabel('Tweet index', color='black', size=15)
plt.ylabel('Publishing time', color='black', size=15)
plt.scatter(data.index.tolist(), data['time'].tolist(), alpha=0.8)

for name in data.screen_name.unique():
    start = data[data.screen_name == name].index[0]
    plt.axvline(start, color='red', linewidth=2, linestyle='--')
    
start = data[data.screen_name == name].index[0]
plt.axvline(start, color='red', label='pages separator', linewidth=2, linestyle='--')
    
plt.legend(loc = 'best')
```
